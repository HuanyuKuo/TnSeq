```python
exps = []
do_exp = 1000
#do_exp = 5000

for run in range(0,do_exp):
    x = Dilution(REPLICA, STRAIN, MEDIA, BCID_REF, NEUTRAL)
    x.read_files()
    #x.get_experimental_s()
    x.runs()
    exps.append(x.s_record)
    if run%100==0:
        print run
exps = np.resize(exps,(do_exp,x.N_BCID))
#stitle = './output/' + REPLICA + '_' + STRAIN + '_' + MEDIA + '_exps_runs_'+ repr(do_exp) + '.txt' # output file name
#np.savetxt(stitle, exps, fmt='%f')
```

```python
# takes time to run s_ave s_var
#do_exp = 1000
#stitle = './output/' + REPLICA + '_' + STRAIN + '_' + MEDIA + '_exps_runs_'+ repr(do_exp) + '.txt' # output file name
#b = np.loadtxt(stitle, dtype=float)
#exps = b
s_arr=[]
s_ave = [0 for i in range(0,x.N_BCID)]
s_var = [0 for i in range(0,x.N_BCID)]
for bc in range(0,x.N_BCID):
    s = []
    for do in range(0,do_exp):
        ss = exps[do][bc]
        if (ss< 100):
            s.append(ss)
    s_arr.append(s)
    if len(s) > 10:
        s_ave[bc] = np.mean(s)
        s_var[bc] = np.var(s)
#print s_var
```

```python
x.get_experimental_s()
#print x.s_inifreq
```

```python
# plot confidence region
xx = np.log10(x.pvals)
yy = np.add(s_ave,2*np.sqrt(s_var)) # up confidence
yy1 = np.subtract(s_ave,2*np.sqrt(s_var)) # down confidence
xx2 = []
xx2.extend(xx[::-1]), xx2.extend(xx)
yy3 = []
yy3.extend(yy1[::-1]),yy3.extend(yy)
fig, ax = plt.subplots(num=None, figsize=(8, 6), dpi=80)
ax.fill(xx2, yy3, color = '0.75')
plt.xlabel("Log10 of initial BC frequency")
plt.ylabel("s confidential region")
s_ref = -1 * x.S_REF * np.ones((len(x.pvals),1)) 
plt.plot(xx,s_ref,'g',label = '- s_Ref')
plt.legend()
# add experimental points # first get the initial freq
exp_ini_freq = x.s_inifreq
exp_s = x.s_slope
exp_s_err = x.s_stderr
'''
filename = './source/' + x.replica + '_' + x.strain + '_' + x.media + '_gen' + repr(10*0) + '.txt' # file name
dat, dat_rows = x.input_dat(filename,3)
bc_count =[]
bc_id = []  
for row in range(1,dat_rows):
    bc_id.append(int(dat[row,0]))
    tmp = [map(float, t) for t in dat[row,1]] 
    bc_count.append(int(dat[row,1]))
            
bc_id = np.array(bc_id) # make format as numpy array
bc_count = np.array(bc_count)  # make format as numpy array
all_count = sum(bc_count)
for i in range(0,len(x.s_bcid)):
    check = False
    for j in range(0,len(bc_id)):
        if bc_id[j] == x.s_bcid[i]:
            exp_ini_freq.append(bc_count[j]/float(all_count))
            exp_s.append(x.s_dat[i])
            exp_s_err.append((1-x.s_r2[i])*0.01/2)
            check = True
            break
    if check == False:
        print x.s_bcid[i]
'''
plt.plot(xx,x.s_record,'r.')
print len(exp_s)/len(x.s_record)
plt.plot(np.log10(exp_ini_freq),exp_s,'b.')
#plt.xlim(-4, -2.5)
#plt.ylim(-0.04, 0.04)
plt.show()
#print exp_s
```
