```python
import numpy as np                 # Math library
import matplotlib.pyplot as plt    # Plots
from scipy.optimize import curve_fit #Curve Fit

from scipy import optimize
from scipy.misc import factorial

xdata = [1,2,3,4,5,6,12]
ydata = [147,47,17,11,8,3,1]


powerlaw = lambda x, amp, index: amp * (x**index)

yerr = [0.2 for y in ydata]         # simulated errors (10%)
logx = np.log10(xdata)
logy = np.log10(ydata)
logyerr = np.divide(yerr, ydata)


# define our (line) fitting function
fitfunc = lambda p, x: p[0] + p[1] * x
errfunc = lambda p, x, y, err: (y - fitfunc(p, x)) / err

pinit = [1.0, -1.0]
out = optimize.leastsq(errfunc, pinit,
                       args=(logx, logy, logyerr), full_output=1)

pfinal = out[0]
covar = out[1]
print pfinal
print covar

index = pfinal[1]
amp = 10.0**pfinal[0]
             
x_fit =  np.linspace(1, 13, 100)
plt.clf()
plt.plot(x_fit, powerlaw(x_fit, amp, index))     # Fit
plt.plot(xdata, ydata, 'ko')  # Data
plt.show()
```
